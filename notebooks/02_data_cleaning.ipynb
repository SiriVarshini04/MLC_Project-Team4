```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/city_day.csv")
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
required_cols = [
    "City", "Date",
    "PM2.5", "PM10", "NO2", "SO2", "CO", "O3",
    "AQI", "AQI_Bucket"
]

df = df[required_cols]
```

```python
df.isnull().sum()
```

```python
pollutants = ["PM2.5", "PM10", "NO2", "SO2", "CO", "O3", "AQI"]

df[pollutants] = df[pollutants].fillna(df[pollutants].median())
```

```python
for col in pollutants:
    df = df[df[col] >= 0]
```

```python
df.drop(columns=["AQI_Bucket"], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.to_csv("../data/processed/aqi_clean.csv", index=False)
print("✅ Cleaned dataset saved successfully")
```

```python
## Data Cleaning Summary

- Selected standard AQI pollutants only
- Removed less significant pollutants with excessive missing values
- Handled missing values using median imputation
- Removed invalid negative pollutant values
- Created a clean dataset for modeling
```
